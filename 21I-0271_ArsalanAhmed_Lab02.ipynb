```python
def fun(num1,num2):
    
    train1=pd.concat([Setosa[:num1],Setosa[num2:]])
    train2=pd.concat([Versicolor[:num1],Versicolor[num2:]])
    train3=pd.concat([Virginica[:num1],Virginica[num2:]])
    test1=Setosa[num1:num2]
    test2=Versicolor[num1:num2]
    test3=Virginica[num1:num2]
    labels=pd.concat([train1["EncodedLabels"],train2["EncodedLabels"],train3["EncodedLabels"]],axis=0)
    labels_Y=pd.concat([test1["EncodedLabels"],test2["EncodedLabels"],test3["EncodedLabels"]],axis=0)
    train_data=pd.concat([train1,train2,train3],axis=0)
    test_data=pd.concat([test1,test2,test3],axis=0)
    random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
    train_data.drop(["EncodedLabels"],axis=1,inplace=True)
    test_data.drop(["EncodedLabels"],axis=1,inplace=True)
    
    x_train=train_data
    y_train=labels
    x_test=test_data
    y_test=labels_Y
    
    random_forest.fit(x_train, y_train)
    
    # Make predictions on the test data
    predictions = random_forest.predict(x_test)

    # Evaluate the model's accuracy
    accuracy = (predictions == y_test).mean()
    print(f"Accuracy: {accuracy:.2f}")
    return accuracy
```
